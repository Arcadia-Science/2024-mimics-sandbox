```r
filter_criteria1 <- read_csv("outputs/viral/all_nomburg_virus_matches_with_metadata_filtered_criteria1.csv", show_col_types = FALSE) %>%
  mutate(host_organism = gsub(" \\(.*", "", host_organism)) %>%
  mutate(host_gene_names_primary = tolower(host_gene_names_primary))
```

```r
unique(filter_criteria1$host_organism)
```

```r
organisms <- unique(filter_criteria1$host_organism)

gene_sets <- lapply(organisms, function(organism) {
  unique(filter_criteria1$host_gene_names_primary[filter_criteria1$host_organism == organism])
})
names(gene_sets) <- organisms

upset_data <- fromList(gene_sets)
```

```r
options(repr.plot.width = 6, repr.plot.height = 3.5, repr.plot.res = 300)

upset(upset_data, 
      nsets = length(organisms),  # Number of sets to include
      order.by = "freq") 
```

```r
filter_criteria1_duplicates_tally <- filter_criteria1 %>%
  group_by(query) %>%
  tally() %>%
  filter(n > 1)

filter_criteria1_duplicates <- filter_criteria1 %>%
  filter(query %in% filter_criteria1_duplicates_tally$query) %>%
  arrange(query)

filter_criteria1_duplicates %>% select(query, host_organism, alntmscore, qtmscore, ttmscore, evalue, host_gene_names_primary)
```

```r
# removing duplicates
filter_criteria1_duplicate_filtered <- filter_criteria1 %>%
  group_by(query) %>%
  slice_min(evalue)
```

```r
# check that retaining the minimum evalue removes all duplicates
nrow(filter_criteria1_duplicate_filtered)
```

```r
organisms <- unique(filter_criteria1_duplicate_filtered$host_organism)

gene_sets <- lapply(organisms, function(organism) {
  unique(filter_criteria1_duplicate_filtered$host_gene_names_primary[filter_criteria1_duplicate_filtered$host_organism == organism])
})
names(gene_sets) <- organisms

upset_data <- fromList(gene_sets)
```

```r
options(repr.plot.width = 6, repr.plot.height = 3.5, repr.plot.res = 300)

upset(upset_data, 
      nsets = length(organisms),  # Number of sets to include
      order.by = "freq") 
```
