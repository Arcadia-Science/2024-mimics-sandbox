```r
setwd("..")
```

```r
# adjust plot size rendered inline
options(repr.plot.width = 6, repr.plot.height = 4, repr.plot.res = 300)
```

```r
curated_pos_ctrls <- read_tsv("~/Downloads/curated-parasite-positive-controls.tsv", show_col_types = FALSE) %>%
  filter(use_as_positive_control == "yes") %>%
  mutate(species = str_remove(pattern = "\\(.*", species),
         species = str_trim(species)) %>%
  arrange(domain, species) %>%
  mutate(species = factor(species, levels = unique(species)))
```

```r
ggplot(curated_pos_ctrls %>% mutate(domain = str_to_sentence(domain)), aes(x = species, fill = domain)) +
  geom_bar() +
  coord_flip() +
  theme_arcadia() +
  scale_fill_arcadia() +
  clean_plot() +
  labs(fill = "Domain") +
  theme(legend.position = "bottom",
        legend.location = "plot") +
  guides(fill=guide_legend(nrow=2, byrow=TRUE))
```
